```python
csv_reader="/Users/kevindorado/Desktop/Correcciones/Copia de global-data-on-sustainable-energy (1).csv"
global_data=pd.read_csv(csv_reader)
global_df=pd.DataFrame(global_data)
global_df.dropna()
global_df
```

```python
import matplotlib.pyplot as plt

# Cambiando el nombre de 'df_analysis' a 'energy_financial_df' para reflejar mejor la información que contiene
energy_financial_df =pd.DataFrame(global_df[['Year', 'Primary energy consumption per capita (kWh/person)', 
                                 'Financial flows to developing countries (US $)']].dropna())

# Agrupar por año y calcular la media de consumo de energía primaria y flujos financieros
annual_averages = energy_financial_df.groupby('Year').mean().reset_index()

# Crear un gráfico de dos ejes para comparar las tendencias
fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('Primary Energy Consumption per Capita (kWh/person)', color=color)
ax1.plot(annual_averages['Year'], annual_averages['Primary energy consumption per capita (kWh/person)'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Crear un segundo eje Y para los flujos financieros
ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Financial Flows to Developing Countries (US $)', color=color)  
ax2.plot(annual_averages['Year'], annual_averages['Financial flows to developing countries (US $)'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  
plt.title('Primary Energy Consumption vs Financial Flows to Developing Countries')
plt.show()

energy_financial_df
```

Interpretación:
Consumo de Energía Primaria per Cápita: Esta métrica refleja la cantidad de energía consumida por persona y es un indicador del nivel de uso de energía en un país. Un aumento en este valor podría indicar un crecimiento económico, un mayor acceso a la energía, o un incremento en la industrialización y la modernización.

Flujos Financieros a Países en Desarrollo: Esto representa la cantidad de dinero, generalmente en forma de ayuda o inversión, que fluye desde países desarrollados hacia países en desarrollo. Un aumento puede ser un indicador de más apoyo internacional, mientras que una disminución puede reflejar una retirada de la inversión o un cambio en las políticas de ayuda exterior.

El gráfico muestra que, durante el período de 2000 a aproximadamente 2013-2014, ambas métricas parecen incrementar, lo cual podría interpretarse como un período de crecimiento y de mayor interacción económica global. A partir de ese punto, el consumo de energía primaria per cápita parece estabilizarse o crecer a un ritmo más lento, mientras que los flujos financieros a países en desarrollo alcanzan un pico y luego caen drásticamente.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Definir nombres de columnas
renewable_capacity_col = 'Renewable-electricity-generating-capacity-per-capita'
renewable_electricity_col = 'Electricity from renewables (TWh)'

# Definir países BRICS + México
brics_countries = ['Brazil', 'Russia', 'India', 'China', 'South Africa', 'Mexico']

# Cargar datos desde CSV
df = pd.read_csv('global-data-on-sustainable-energy (1).csv')

# Filtrar por países BRICS + México
df_filtered = df[df['Entity'].isin(brics_countries)]

# Agrupar por país y obtener el último año para cada país
df_grouped = df_filtered.groupby('Entity').tail(1)

# Filtrar para valores no nulos
df_grouped = df_grouped.dropna(subset=[renewable_capacity_col, renewable_electricity_col])

# Obtener países
countries = df_grouped['Entity'].to_numpy()

# Obtener datos para cada columna
capacity_data = df_grouped[renewable_capacity_col].to_numpy()
electricity_data = df_grouped[renewable_electricity_col].to_numpy()

# Configurar el gráfico
plt.figure(figsize=(12, 8))
bar_width = 0.35
index = range(len(countries))

# Crear barras apiladas
plt.bar(index, capacity_data, bar_width, label=renewable_capacity_col)
plt.bar(index, electricity_data, bar_width, bottom=capacity_data, label=renewable_electricity_col)

# Establecer etiquetas y título
plt.xlabel('Countries')
plt.ylabel('Renewable electricity generated per capita')
plt.xticks(index, countries, rotation=45, ha='right')
plt.title('Comparison of Renewable Electricity (BRICS + Mexico)')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()

```

El gráfico de barras apiladas muestra la capacidad de generación de electricidad renovable per cápita (kW/persona) y la electricidad renovable total generada (TWh) para los países BRICS y México. Se observan las siguientes tendencias:

1. Capacidad de generación de electricidad renovable per cápita:

    Brasil tiene la mayor capacidad de generación per cápita (alrededor de 3.5 kW/persona), seguido de México (alrededor de 1.2 kW/persona).
    China y Sudáfrica tienen capacidades similares, alrededor de 0.8 kW/persona.

2. Electricidad renovable total generada:

    Brasil se ubica en tercer lugar, con una generación de alrededor de 500 TWh.
    Rusia y México generan cantidades similares, alrededor de 200 TWh y 180 TWh, respectivamente.
